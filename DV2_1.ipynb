```python
import numpy as np
import matplotlib.pyplot as plt

# Grid size
nx, ny = 20, 20

# Create coordinate grid
x = np.linspace(0, 2 * np.pi, nx)
y = np.linspace(0, 2 * np.pi, ny)
X, Y = np.meshgrid(x, y)

# Scalar field: A scalar function like temperature or elevation
scalar_field = np.sin(X) * np.cos(Y)

# Vector field: A vector function like velocity or wind
U = np.cos(Y)  # x-component
V = np.sin(X)  # y-component

# Create the figure and subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot scalar field using imshow (or pcolormesh/contourf)
c1 = ax[0].imshow(scalar_field, extent=[0, 2*np.pi, 0, 2*np.pi],
                  origin='lower', cmap='viridis', aspect='auto')
ax[0].set_title('Scalar Field (e.g., Temperature)')
fig.colorbar(c1, ax=ax[0])

# Plot vector field using quiver
ax[1].quiver(X, Y, U, V, color='r')
ax[1].set_title('Vector Field (e.g., Wind Velocity)')
ax[1].set_xlim([0, 2*np.pi])
ax[1].set_ylim([0, 2*np.pi])

plt.tight_layout()
plt.show()
```
