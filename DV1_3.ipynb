```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset from URL
data_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(data_url)
# Create subplots with increased figure size
fig, axes = plt.subplots(2, 4, figsize=(32, 16))  # Increased size and arranged in 2 rows
# Bar Chart
df['species'].value_counts().plot(kind='bar', color='skyblue', ax=axes[0, 0])
axes[0, 0].set_title('Bar Chart of Species', fontsize=14)
axes[0, 0].set_xlabel('Species', fontsize=12)
axes[0, 0].set_ylabel('Count', fontsize=12)
axes[0, 0].tick_params(axis='x', rotation=45)
# Histogram
sns.histplot(df['sepal_length'], bins=20, kde=True, color='purple', ax=axes[0, 1])
axes[0, 1].set_title('Histogram of Sepal Length', fontsize=14)
axes[0, 1].set_xlabel('Sepal Length', fontsize=12)
axes[0, 1].set_ylabel('Frequency', fontsize=12)
# Box Plot
sns.boxplot(x=df['sepal_width'], color='green', ax=axes[0, 2])
axes[0, 2].set_title('Box Plot of Sepal Width', fontsize=14)
axes[0, 2].set_xlabel('Sepal Width', fontsize=12)
# Pie Chart
df['species'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'red', 'green'], ax=axes[0, 3])
axes[0, 3].set_title('Pie Chart of Species', fontsize=14)
axes[0, 3].set_ylabel('')  # Hide the ylabel for aesthetics
# Scatter Plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', palette='viridis', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Scatter Plot of Sepal Dimensions', fontsize=14)
# Violin Plot
sns.violinplot(x='species', y='petal_length', data=df, palette='pastel', ax=axes[1, 1])
axes[1, 1].set_title('Violin Plot of Petal Length by Species', fontsize=14)
# Heatmap
numerical_df = df.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5, ax=axes[1, 2])
axes[1, 2].set_title('Heatmap of Feature Correlation', fontsize=14)
# Pair Plot (Cannot be directly included in subplots, so shown separately)
sns.pairplot(df, hue='species', palette='dark')
plt.suptitle('Pair Plot of Iris Features', fontsize=16)
# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
